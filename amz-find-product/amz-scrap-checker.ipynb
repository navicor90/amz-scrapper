## Scrapping and saving soups

```python
driver = webdriver.Firefox()
```

```python
import urllib
base_url = "https://www.amazon.es"
    
#soups.append(soup_content_from_url(driver, url))
```

```python
last_index = 120
```

```python
import time
import random

def search_in_amz_with_mcrdf(mcrdf, last_index = 0, limit=100):
    def search_in_amz(keywords):
        params={"k":urllib.parse.quote(keywords),
                "__mk_es_ES":"ÅMÅŽÕÑ",
                "ref":"nb_sb_noss"}

        def create_params(d):
            r = ""
            for k in d.keys():
                r = r+f"{k}={d[k]}&"
            return r[:-1]

        url = base_url+'/s?'+create_params(params)
        return soup_content_from_url(driver,url)
        
    i = last_index
    limit = limit + i
    for i,r in mcrdf[mcrdf.index > last_index].iterrows():
        time.sleep(random.randint(1,5))
        if i > limit:
            break
        print(str(i)+" : "+str(r['ref.mcr'])+"  "+r['part number'])
        keywords = r['part number']
        s = search_in_amz(keywords)
        file_name = str(r['ref.mcr'])
        save_soup(s, file_name)
    return i-1

last_index = search_in_amz_with_mcrdf(df_5, last_index,100)
last_index
```

## Working with soups

```python
from bs4 import BeautifulSoup

def read_soup(filename):
    with open("soups/"+filename+".html", "r") as f:
        contents = f.read()
        soup = BeautifulSoup(contents, 'lxml')
        return soup
```

```python
from os import listdir
files = listdir('soups')
ss = {}
for f in files:
    ref_mcr = f.replace('.html','')
    ss[ref_mcr] = read_soup(ref_mcr)
```

```python
from utils import amz_products_with_soup,ProductSearch
products = []
for ref,soup in ss.items():
    products.append(ProductSearch('MCR',ref,amz_products_with_soup(soup),soup))
len(products)
```

```python
interesting_products = []
for p in products:
    if len(p.to_list()) == 0:
        interesting_products.append(int(p.ref_local_id))
```

```python
ordered = df[df['ref.mcr'].isin(interesting_products)].sort_values(by='precio',ascending=False)
ordered.head(20)
```
