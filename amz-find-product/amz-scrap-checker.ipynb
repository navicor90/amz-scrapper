## Scrapping and saving soups

```python
driver = webdriver.Firefox()
```

```python
import urllib
base_url = "https://www.amazon.es"
    
#soups.append(soup_content_from_url(driver, url))
```

```python
last_index = 220
```

```python
import time
import random

def search_in_amz_with_mcrdf(mcrdf, last_index = 0, limit=100):
    def search_in_amz(keywords):
        params={"k":urllib.parse.quote(keywords),
                "__mk_es_ES":"ÅMÅŽÕÑ",
                "ref":"nb_sb_noss"}

        def create_params(d):
            r = ""
            for k in d.keys():
                r = r+f"{k}={d[k]}&"
            return r[:-1]

        url = base_url+'/s?'+create_params(params)
        return soup_content_from_url(driver,url)
        
    i = last_index
    limit = limit + i
    for i,r in mcrdf[mcrdf.index > last_index].iterrows():
        time.sleep(random.randint(1,5))
        if i > limit:
            break
        print(str(i)+" : "+str(r['ref.mcr'])+"  "+r['part number'])
        keywords = r['part number']
        s = search_in_amz(keywords)
        file_name = str(r['ref.mcr'])
        save_soup(s, file_name)
    return i-1

last_index = search_in_amz_with_mcrdf(df_5, last_index,100)
last_index
```

## Working with soups

```python
from amz_search_utils import read_soup
```

```python
from amz_search_utils import AmazonProductSearch
from os import listdir

files = listdir('soups')
amazon_searchs = {}
for f in files:
    ref_mcr = f.replace('.html','')
    soup = read_soup(ref_mcr)
    amazon_searchs[ref_mcr] = AmazonProductSearch(soup)

print(len(files))
print(len(amazon_searchs))
```

```python
items = []
for ref_id, amazon_search in amazon_searchs.items():
    for search_item in amazon_search.searchItems():
        item = search_item.to_dict()
        item['ref_id'] = ref_id
        items.append(item)
print(len(items))
print(items[0])
```

```python
save_soup(amazon_searchs['1321675'].soup,'test')
```

```python
dfamz = pd.DataFrame(items)
dfamz.head(2)
```

### Founded ref ids
When we search in amazon, maybe the amazon results are empty. So in "found_ref_ids" var we have just those which exists in amazon.

```python
dfamz = dfamz[dfamz.ref_id != 'test']
found_ref_ids = list(dfamz.ref_id.astype(int).unique())
len(found_ref_ids)
```

### Amazon searched Ids
They are all those we look for in amazon, including if they doesn't have a search result.

```python
amz_searched_ids = [int(k) for k in amazon_searchs.keys() if k!='test']
len(amz_searched_ids)
```

### Filtering MCR DataFrame
We filter MCR dataframe looking for products that not exist in amazon

```python
searched_df = df[df['ref.mcr'].isin(amz_searched_ids)].sort_values(by='precio',ascending=False)
not_in_amz_df = searched_df[~searched_df['ref.mcr'].isin(found_ref_ids)].sort_values(by='precio',ascending=False)
not_in_amz_df = not_in_amz_df[not_in_amz_df.stock > 5]
print(f'Length {len(not_in_amz_df)}')
not_in_amz_df.head(20)#[ordered.precio < 1000].head(20)
```
